## Load CSV

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
df = pd.read_csv("../Output/rata-rata_per_ton.csv")
```

```python
top10 = df.sort_values("rata_rata_biaya_per_ton_pit", ascending=False).head(10)
```

```python
top10 = top10.sort_values("rata_rata_biaya_per_ton_pit")
```

```python
os.makedirs("assets", exist_ok=True)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top10["pit_id"].astype(str), top10["rata_rata_biaya_per_ton_pit"])
plt.xlabel("Pit ID")
plt.ylabel("Rata Rata Biaya Per Ton")
plt.title("Top 10 Rata Rata Biaya Per Ton per Pit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("assets/top10_biaya_per_pit.png", dpi=300)

plt.show()
```

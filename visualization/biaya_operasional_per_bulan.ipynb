## Load CSV

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Output/biaya_operasional_per_bulan.csv")

df["bulan"] = pd.to_datetime(df["bulan"])
```

```python
df["moving_avg"] = df["total_biaya"].rolling(window=3).mean()

plt.figure(figsize=(12,6))
plt.plot(df["bulan"], df["total_biaya"], label="Total Biaya")
plt.plot(df["bulan"], df["moving_avg"], label="Moving Avg 3 Bulan")
plt.xlabel("Bulan")
plt.ylabel("Total Biaya Operasional")
plt.title("Tren Bulanan Biaya Operasional + 3 Bulan Moving Avg")
plt.legend()
plt.grid(True)
plt.tight_layout()

# simpan gambar
plt.savefig("assets/tren_biaya_operasional.png", dpi=300)

plt.show()
```

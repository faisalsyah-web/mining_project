## Load CSV

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Output/biaya_operasional_per_bulan(kategori).csv")

# Pastikan kolom bulan sebagai datetime (opsional tapi penting untuk sort)
df["bulan"] = pd.to_datetime(df["bulan"], format="%Y-%m")
```

```python
# Hitung total biaya per kategori per bulan
pivot_df = df.pivot_table(
    index="bulan",
    columns="kategori",
    values="total_biaya",
    aggfunc="sum")
```

```python
# Plot
plt.figure(figsize=(14,6))
for col in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[col], label=col)

plt.xlabel("Bulan")
plt.ylabel("Total Biaya Operasional")
plt.title("Tren Biaya Operasional per Kategori per Bulan")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Simpan gambar
plt.savefig("assets/tren_biaya_per_kategori.png", dpi=300)

plt.show()
```
